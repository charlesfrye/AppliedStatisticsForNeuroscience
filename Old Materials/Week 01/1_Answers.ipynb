## Question 2: What's in a `norm`?

Explain the `dnorm`, `pnorm`, and `qnorm` functions in your own words.

```r
qnorm(0.73)
```

`dnorm` takes in a number (and, optionally, the parameters of a normal distribution) and then returns the height of the probability density function (PDF) of a normal random variable with those parameters at that value. If you were to generate lots of examples from `rnorm`, create a very fine histogram, and look at a skinny histogram box around the number given to `dnorm`, the height would be close to the value given by `dnorm`.

`pnorm` takes in a number (and, optionally, the parameters of a normal distribution) and then returns the value of the cumulative density function of the normal random variable with those parameters at that value. If you were to generate lots of examples from `rnorm`, create a very fine histogram, and then add up the areas of all of the histogram boxes with right edges below the number given to `pnorm`, the sum total of those heights would be close to the result given by `pnorm`.

Note that the above examples assume that you create a *normalized* histogram, that is, a histogram where the total area adds up to 1. You can make a normalized histogram by calling `hist` with the optional argument `freq=FALSE`. This means that the results of `pnorm` are always less than 1. Is the same true for `dnorm`?

`qnorm` is the inverse of `pnorm`. It takes in a value (call it $P$) between 0 and 1 and returns a number. If you were to calculate the $100*P$th percentile of a large enough sample provided by `rnorm`, the value should be very close to the result given by `qnorm`.

#### What's a probability density function?

Probability density functions (PDFs) get their name because they are a lot like the physical notion of "density". A __density function__ for some material tells you how much mass is in a given length (for a 1-D object, like a rope or a wire), area (for a 2-D object, like a sheet) or volume (for a 3-D object, like a hunk of metal or the brain). A __probability density function__ tells you how much *probability* is in a given region. These regions can be 1-D, as they are for a Gaussian random variable, or they can be volumes, as they would be if you were describing the probability of finding an object at some point in a 3-D space, or they can be more exotic regions, like higher-dimensional volumes, as when you describe the probability density function of the firing rates of a population of neurons.

#### What's a cumulative density function?

Cumulative density functions (CDFs) tell you how much probability has "accumulated" up to a given point. Formally, they tell you the probability that a random variable has a value that is less than or equal to the value you put into the CDF. As an example, because Gaussians are symmetric, the value of their CDF is 0.5 at their mean -- there's a 50% chance that a random variable that is distributed according to a Gaussian is below its mean.

Youâ€™re studying feeding behavior in rats and you want to know the average amount of food, in grams, that the members of your rat colony eat per day. You (or, more accurately, your undergrad) measures the food intake of each rat, resulting in the values below.

First, calculate the mean, median, and standard deviation of your sample. Produce a histogram with six bins and six axis labels.
Then, create a function that computes a 95% confidence interval from an input vector and calculate a 95% confidence interval on the data. 
Finally, explain, in your own words, what a confidence interval is. Answer in the text box below.

```r
ratData <- c(   31,15,12,21,26,10,12,25,27,25,
                16,19,16,18,8,13,22,19,29,17,
                30,14,9,22,28,19,30,23,20,17)

ratMean <- mean(ratData)
ratMedian <- median(ratData)
ratSD <- sqrt(var(ratData))

err <- qnorm(0.975)*(ratSD)/sqrt(length(ratData))

confInterval <- c(ratMean-err,ratMean+err)

myConfFunk <- function(x) {
    N <- length(x)
    xMean <- mean(x); xSD <- sqrt(var(x))
    xErr <- qnorm(0.975)*(xSD)/sqrt(N)
    return(c(xMean-xErr,xMean+xErr))
}

print(confInterval)
print(myConfFunk(ratData))
```

If we were to run our experiment 100 times and compute 100 confidence intervals at a confidence level of X% for our parameter, X of the confidence intervals computed from the data would contain the true value of the parameter. Sometimes, this statement is shortened to "there is an X% chance that the true value of the parameter is in this interval", but this phrasing is misleading. For example, the chance that the mean of a second, independent and identical experiment is within the 95% confidence interval is actually lower -- sometimes much lower!
